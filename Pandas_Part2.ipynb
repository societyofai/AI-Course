# How to read a data file in pandas?

```python
drinks=pd.read_table('http://bit.ly/drinksbycountry')
drinks.head()
```

### why the commas? It looks wierd right? 
It is comma seprated.

```python
drinks=pd.read_table('http://bit.ly/drinksbycountry',sep=',')
drinks.head()
```

### Or we can use the other method 

```python
drinks=pd.read_csv('http://bit.ly/drinksbycountry')  # use comma as default seprator
drinks.columns
```

# How to sort a dataframe by a coulmn(series)

### In ascending order

```python
drinks.sort_values('continent',inplace=True)
drinks
```

```python
#sorting by multiple parameters
drinks.sort_values(['continent','total_litres_of_pure_alcohol'],inplace=True)
drinks.head()
```

### in decending order

```python
drinks.sort_values('continent',ascending=False)
```
